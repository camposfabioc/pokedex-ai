## Pytorch Data Preparation

```python
from torchvision.transforms import v2
```

```python
train_transforms = v2.Compose([
    v2.RandomHorizontalFlip(p=0.5),
    v2.RandomRotation(10),
    v2.ColorJitter(brightness=0.2, contrast=0.2, saturation=0.2),
    v2.ToTensor(),
    v2.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
])

val_test_transforms = v2.Compose([
    v2.ToTensor(),
    v2.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
])
```

```python
from torchvision.datasets import ImageFolder

train_dataset = ImageFolder(root="../data/splits/train", transform=train_transforms)
val_dataset = ImageFolder(root="../data/splits/validation", transform=val_test_transforms)
test_dataset = ImageFolder(root="../data/splits/test", transform=val_test_transforms)

print(f"Train: {len(train_dataset)} imagens")
print(f"Val: {len(val_dataset)} imagens")
print(f"Test: {len(test_dataset)} imagens")
print(f"Classes: {len(train_dataset.classes)}")
```

```python
from torch.utils.data import DataLoader

train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True, num_workers=0)
val_loader = DataLoader(val_dataset, batch_size=32, shuffle=False, num_workers=0)
test_loader = DataLoader(test_dataset, batch_size=32, shuffle=False, num_workers=0)
```

Now the data pipeline is done, I will run a quick check to validate whether it is working

```python

images, labels = next(iter(train_loader))

print(f"Batch shape: {images.shape}")
print(f"Labels shape: {labels.shape}")
print(f"Primeiro label: {labels[0]}")
```

And also visually check the augmentation:

```python
def denormalize(tensor):
    mean = np.array([0.485, 0.456, 0.406])
    std = np.array([0.229, 0.224, 0.225])
    img = tensor.numpy().transpose(1, 2, 0)
    img = std * img + mean
    img = np.clip(img, 0, 1)
    return img

plt.imshow(denormalize(images[0]))
plt.title(f"Label: {labels[0]}")
plt.show()
```

## Model training (Transfer Learning)

Since the main idea of this project is to have it deployed in an actual software (RaspberryPi) and run image classifications on actual Pokemon pictures. I'm choosing the MobilNetV2 model because it is small enough to fit and the performance is the best we can get with this portability.

```python
import torch
from torchvision import models
import torch.nn as nn

model = models.mobilenet_v2(weights='IMAGENET1K_V1')

# Freezing parmetros
for param in model.parameters():
    param.requires_grad = False

#Replace last layer
model.classifier[1] = nn.Linear(model.classifier[1].in_features, 150)
```

```python
device = torch.device("cpu")
model = model.to(device)
```

```python
criterion = nn.CrossEntropyLoss()

optimizer = torch.optim.Adam(model.classifier[1].parameters(), lr=0.001)
```

```python
from tqdm import tqdm

num_epochs = 5

for epoch in range(num_epochs):
    print(f"Epoch {epoch+1}/{num_epochs}")

    model.train()
    train_loss = 0.0
    train_correct = 0

    for image,label in tqdm(train_loader, desc="Training"):
        image = image.to(device)
        label = label.to(device)
    
        #Forward pass
        outputs = model(image)
        loss = criterion(outputs, label)

        #Backward pass
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        #metrics
        train_loss += loss.item()
        _, predicted = torch.max(outputs,1)
        train_correct += (predicted == label).sum().item()

    train_loss = train_loss/len(train_loader)
    train_acc = train_correct/len(train_dataset)

    #Validation
    model.eval()
    val_correct = 0

    with torch.no_grad():
        for image, label in tqdm(val_loader, desc="Validation"):
            image = image.to(device)
            label = label.to(device)

            outputs = model(image)
            _, predicted = torch.max(outputs, 1)
            val_correct += (predicted == label).sum().item()
    
    val_acc = val_correct / len(val_dataset)

    print(f"Train Loss: {train_loss:.4f}, Train Acc: {train_acc:.4f}, Val Acc: {val_acc:.4f}\n")
```

I will train 5 more epochs to see if the performance on the validation set improves more.

```python
total_epochs_trained = 5

num_epochs = 5

for epoch in range(num_epochs):
    current_epoch = total_epochs_trained + epoch + 1
    print(f"Epoch {current_epoch}/{total_epochs_trained + num_epochs}")

    model.train()
    train_loss = 0.0
    train_correct = 0

    for image,label in tqdm(train_loader, desc="Training"):
        image = image.to(device)
        label = label.to(device)
    
        #Forward pass
        outputs = model(image)
        loss = criterion(outputs, label)

        #Backward pass
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        #metrics
        train_loss += loss.item()
        _, predicted = torch.max(outputs,1)
        train_correct += (predicted == label).sum().item()

    train_loss = train_loss/len(train_loader)
    train_acc = train_correct/len(train_dataset)

    #Validation
    model.eval()
    val_correct = 0

    with torch.no_grad():
        for image, label in tqdm(val_loader, desc="Validation"):
            image = image.to(device)
            label = label.to(device)

            outputs = model(image)
            _, predicted = torch.max(outputs, 1)
            val_correct += (predicted == label).sum().item()
    
    val_acc = val_correct / len(val_dataset)

    print(f"Train Loss: {train_loss:.4f}, Train Acc: {train_acc:.4f}, Val Acc: {val_acc:.4f}\n")
```

After epoch 7, the accuracy started decreasing... I will now test the performance increasing from modifying only the last layer to the last 4 layers.

```python
model = models.mobilenet_v2(weights='IMAGENET1K_V1')

for param in model.parameters():
    param.requires_grad = False

model.classifier[1] = nn.Linear(model.classifier[1].in_features, 150)

# Unfreeze last 4 layers
for param in model.features[-4:].parameters():
    param.requires_grad = True

model = model.to(device)

optimizer = torch.optim.Adam([
    {'params': model.features[-4:].parameters(), 'lr': 0.0001},
    {'params': model.classifier[1].parameters(), 'lr': 0.001}
], lr=0.001)
```

```python
num_epochs = 10

for epoch in range(num_epochs):
    print(f"Epoch {epoch+1}/{num_epochs}")

    model.train()
    train_loss = 0.0
    train_correct = 0

    for image,label in tqdm(train_loader, desc="Training"):
        image = image.to(device)
        label = label.to(device)
    
        #Forward pass
        outputs = model(image)
        loss = criterion(outputs, label)

        #Backward pass
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        #metrics
        train_loss += loss.item()
        _, predicted = torch.max(outputs,1)
        train_correct += (predicted == label).sum().item()

    train_loss = train_loss/len(train_loader)
    train_acc = train_correct/len(train_dataset)

    #Validation
    model.eval()
    val_correct = 0

    with torch.no_grad():
        for image, label in tqdm(val_loader, desc="Validation"):
            image = image.to(device)
            label = label.to(device)

            outputs = model(image)
            _, predicted = torch.max(outputs, 1)
            val_correct += (predicted == label).sum().item()
    
    val_acc = val_correct / len(val_dataset)

    print(f"Train Loss: {train_loss:.4f}, Train Acc: {train_acc:.4f}, Val Acc: {val_acc:.4f}\n")
```

Now the performance is quite good on the validation set (91.8%), I will proceed with some error understanding

```python
model.eval()
errors = []

with torch.no_grad():
    for images, labels in val_loader:
        images_device = images.to(device)
        outputs = model(images_device)
        _, predicted = torch.max(outputs, 1)
        
        wrong_idx = (predicted != labels.to(device)).nonzero().squeeze()
        
        if wrong_idx.numel() > 0:
            for idx in wrong_idx:
                errors.append({
                    'image': images[idx],
                    'true': val_dataset.classes[labels[idx]],
                    'pred': val_dataset.classes[predicted[idx]]
                })
        
        if len(errors) >= 9:
            break

fig, axes = plt.subplots(3, 3, figsize=(12, 12))
for i, ax in enumerate(axes.flat):
    if i < len(errors):
        img = errors[i]['image'].permute(1, 2, 0).numpy()
        img = img * [0.229, 0.224, 0.225] + [0.485, 0.456, 0.406]
        img = np.clip(img, 0, 1)
        ax.imshow(img)
        ax.set_title(f"True: {errors[i]['true']}\nPred: {errors[i]['pred']}", fontsize=8)
        ax.axis('off')
plt.tight_layout()
plt.show()
```

```python
from sklearn.metrics import classification_report

model.eval()
all_preds = []
all_labels = []

with torch.no_grad():
    for images, labels in val_loader:
        images = images.to(device)
        outputs = model(images)
        _, predicted = torch.max(outputs, 1)
        
        all_preds.extend(predicted.cpu().numpy())
        all_labels.extend(labels.cpu().numpy())


report = classification_report(all_labels, all_preds, target_names=val_dataset.classes, output_dict=True)

class_scores = {pokemon: report[pokemon]['recall'] for pokemon in val_dataset.classes}

sorted_scores = sorted(class_scores.items(), key=lambda x: x[1])

print("10 Pokmon with worst performance on validation set:")
for pokemon, score in sorted_scores[:10]:
    print(f"{pokemon}: {score:.2%}")
```

## Model testing

```python
model.eval()
test_correct = 0
test_preds = []
test_labels = []

with torch.no_grad():
    for images, labels in tqdm(test_loader, desc="Testing"):
        images = images.to(device)
        labels = labels.to(device)
        
        outputs = model(images)
        _, predicted = torch.max(outputs, 1)
        
        test_correct += (predicted == labels).sum().item()
        test_preds.extend(predicted.cpu().numpy())
        test_labels.extend(labels.cpu().numpy())

test_acc = test_correct / len(test_dataset)
print(f"\n{'='*50}")
print(f"Test accuracy: {test_acc:.4f}")
print(f"{'='*50}\n")

from sklearn.metrics import classification_report

print(classification_report(test_labels, test_preds, target_names=test_dataset.classes, zero_division=0))
```

The model showed a good generalization without overfitting, with the test accuracy (91.4%) very close to the validation accuracy.
